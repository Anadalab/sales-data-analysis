# 📊 Sales Data Analysis
Análisis exploratorio de ventas usando Python (Pandas + Seaborn).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar dataset
df = pd.read_csv('../data/raw/sales_data.csv')
df.head()
```

```python
# Limpieza básica
df.dropna(inplace=True)
df.describe()
```

```python
# Ventas totales por día
sales_by_day = df.groupby('day')['total_bill'].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_day.index, y=sales_by_day.values)

plt.title('Ventas totales por día')
plt.xlabel('Día')
plt.ylabel('Ingresos')

plt.savefig('../images/sales_by_day.png')
plt.show()
```
